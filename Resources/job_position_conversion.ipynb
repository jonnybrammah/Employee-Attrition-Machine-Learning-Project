```python
import pandas as pd
from pathlib import Path
```

```python
#csv_path = "Project_4/Bedbug_Reporting.csv"
glassdoor_data_df = pd.read_csv("clean_glasss_door.csv")
glassdoor_data_df.head()
```

```python
 df = glassdoor_data_df.dropna()
```

```python
df.columns
```

```python
for i, genre in enumerate(df['current']):
    if genre == 'Former Employee':
        df.loc[i, 'employee_position'] = 1
    elif genre == 'Former Employee, less than 1 year':
        df.loc[i, 'employee_position'] = 2
    elif genre == 'Former Employee, more than 1 year':
        df.loc[i, 'employee_position'] = 3
    elif genre == 'Former Employee, more than 3 years':
        df.loc[i, 'employee_position'] = 4
    elif genre == 'Former Employee, more than 5 years':
        df.loc[i, 'employee_position'] = 5
    elif genre == 'Former Employee, more than 8 years':
        df.loc[i, 'employee_position'] = 6
    elif genre == 'Former Employee, more than 10 years':
        df.loc[i, 'employee_position'] = 7
    elif genre == 'Current Employee':
        df.loc[i, 'employee_position'] = 8
    elif genre == 'Current Employee, less than 1 year':
        df.loc[i, 'employee_position'] = 9
    elif genre == 'Current Employee, more than 1 year':
        df.loc[i, 'employee_position'] = 10
    elif genre == 'Current Employee, more than 3 year':
        df.loc[i, 'employee_position'] = 11
    elif genre == 'Current Employee, more than 5 year':
        df.loc[i, 'employee_position'] = 12
    elif genre == 'Current Employee, more than 8 year':
        df.loc[i, 'employee_position'] = 13
    elif genre == 'Current Employee, more than 10 year':
        df.loc[i, 'employee_position'] = 14
print(df['employee_position'])
```
