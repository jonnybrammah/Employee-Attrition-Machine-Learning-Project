```python
import pandas as pd
```

### BLS Quit and Layoff Data

```python
quits_and_layoffs = "../Resources/BLS Data/BLS_Quit_and_Layoff_Data_2001_onwards.csv"
quits_and_layoffs_df = pd.read_csv(quits_and_layoffs, encoding="ISO-8859-1")
```

```python
quits_and_layoffs_df.head(10)
```

```python
#Rename the columns
quits_and_layoffs_df = quits_and_layoffs_df.rename(columns = {"ï»¿FRED Graph Observations": "Date",
                                                             "Unnamed: 1": "Quits (1000s)",
                                                             "Unnamed: 2": "Layoffs (1000s)"})

quits_and_layoffs_df.head()
```

```python
#Drop the first 12 rows since they contain information about the data
quits_and_layoffs_df = quits_and_layoffs_df.iloc[11:].reset_index(drop = True)
```

```python
quits_and_layoffs_df.head()
```

```python
quits_and_layoffs_df.dtypes
```

```python
#Convert the date column to a date type

quits_and_layoffs_df['Date'] = pd.to_datetime(quits_and_layoffs_df['Date'])
quits_and_layoffs_df.head()
```

```python
quits_and_layoffs_df["Quits (1000s)"] = quits_and_layoffs_df["Quits (1000s)"].astype(float)
quits_and_layoffs_df["Layoffs (1000s)"] = quits_and_layoffs_df["Layoffs (1000s)"].astype(float)
```

```python
quits_and_layoffs_df.dtypes
```

```python
quits_and_layoffs_df.to_csv('../Resources/Cleaned BLS Data/BLS_Quit_and_Layoff_Data_2001_onwards_Clean.csv', index=False)
```
